### Шаг 3: скачай результаты

После завершения — скопируй CSV файлы в `/kaggle/working/` чтобы скачать их через Kaggle UI.

```python
import shutil
results_src = Path("/kaggle/working/nlp_case_study/results")
results_dst = Path("/kaggle/working/results")
results_dst.mkdir(exist_ok=True)

for f in results_src.glob("*.csv"):
    shutil.copy(f, results_dst / f.name)
    print(f"Copied {f.name}")

poster_src = Path("/kaggle/working/nlp_case_study/poster")
poster_dst = Path("/kaggle/working/poster")
poster_dst.mkdir(exist_ok=True)
for f in poster_src.glob("*.png"):
    shutil.copy(f, poster_dst / f.name)
    print(f"Copied {f.name}")

print("\nDone! Download from Output tab on the right.")
```
